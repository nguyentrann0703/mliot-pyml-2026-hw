## Xây dựng mô hình Logistic Regression và K-Nearest Neighbors (KNN) trên tập Dry Bean Dataset

## 1. Import thư viện

```python
import pandas as pd

from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier


from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    confusion_matrix
)

pd.set_option("display.max_columns", None)
```

## 2. Đọc dữ liệu

```python
from pathlib import Path
import pandas as pd

DATA_DIR = Path("Dry_Bean_Dataset")

TRAIN_PATH = DATA_DIR / "dry_bean_train.csv"
TEST_PATH = DATA_DIR / "dry_bean_test.csv"

print("Train:", TRAIN_PATH.resolve())
print("Test :", TEST_PATH.resolve())

if not TRAIN_PATH.exists():
    raise FileNotFoundError(f"Không tìm thấy file tại: {TRAIN_PATH.resolve()}")

if not TEST_PATH.exists():
    raise FileNotFoundError(f"Không tìm thấy file tại: {TEST_PATH.resolve()}")

train_df = pd.read_csv(TRAIN_PATH)
test_df = pd.read_csv(TEST_PATH)

print("\nKích thước tập Train:", train_df.shape)
print("Kích thước tập Test :", test_df.shape)

display(train_df.head())
```

## 3. Tách feature và target

```python
X_train = train_df.drop(columns="class")
y_train = train_df["class"]

X_test = test_df.drop(columns="class")
y_test = test_df["class"]

print(y_train.unique())
```

## 4. Logistic Regression

```python
logistic_model = Pipeline([
    ("scaler", StandardScaler()),
    ("classifier", LogisticRegression(max_iter=1000, random_state=42))
])

logistic_model.fit(X_train, y_train)
```

## 5. Đánh giá Logistic Regression

```python
y_pred_lr = logistic_model.predict(X_test)

accuracy_lr = accuracy_score(y_test, y_pred_lr)
precision_lr = precision_score(y_test, y_pred_lr, average="weighted")
recall_lr = recall_score(y_test, y_pred_lr, average="weighted")
f1_score_lr = f1_score(y_test, y_pred_lr, average="weighted")

lr_metrics = pd.DataFrame({
    "Accuracy": [accuracy_lr],
    "Precision": [precision_lr],
    "Recall": [recall_lr],
    "F1 score" : [f1_score_lr]},
    index = ["Logistic Regression"]
                           )

display(lr_metrics)
print("Confusion Matrix".center(30,'-'))
print(confusion_matrix(y_test, y_pred_lr))
```

## 6. K-Nearest Neighbors (KNN)

```python
knn_model = Pipeline([
    ("scaler", StandardScaler()),
    ("classifier", KNeighborsClassifier(n_neighbors=5))
])

knn_model.fit(X_train, y_train)
```

## 7. Đánh giá KNN

```python
y_pred_knn = knn_model.predict(X_test)

accuracy_knn = accuracy_score(y_test, y_pred_knn)
precision_knn = precision_score(y_test, y_pred_knn, average="weighted")
recall_knn = recall_score(y_test, y_pred_knn, average="weighted")
f1_score_knn = f1_score(y_test, y_pred_knn, average="weighted")

knn_metrics = pd.DataFrame({
    "Accuracy": [accuracy_knn],
    "Precision": [precision_knn],
    "Recall": [recall_knn],
    "F1 score" : [f1_score_knn]},
    index = ["KNN (k=5)"]
                           )

display(knn_metrics)
print("Confusion Matrix".center(30,'-'))
print(confusion_matrix(y_test, y_pred_knn))
```

# 8. So sánh mô hình Logistic Regresssion và K-Nearest Neighbors

```python
result = pd.DataFrame({
    "Model": ["Logistic Regression", "KNN (k=5)"],
    "Accuracy": [accuracy_lr, accuracy_knn],
    "Precision": [precision_lr, precision_knn],
    "Recall": [recall_lr, recall_knn],
    "F1 score" : [f1_score_lr, f1_score_knn]
})
display(result)
```

## Nhận xét

- Cả Logistic Regression và KNN đều đạt trên 91% ở cả 4 metrics là `accuracy`, `precision`, `recall` và `f1 score`
- Logistic Regression cho `accuracy` 91.92%, cao hơn KNN (91.55%).
- Precision, Recall và F1-score của Logistic Regression cũng đều cao hơn KNN.
Do đó Logistic Regression là mô hình phù hợp hơn cho bộ dữ liệu Dry Bean trong thí nghiệm này
